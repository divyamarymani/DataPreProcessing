```python
import pandas as pd #data manupulation
import numpy as np #numerical calculation
import matplotlib.pyplot as plt  #data visualization
import seaborn as sns  #advanced data visualization
```

```python
data = pd.read_csv("iris.csv")
data.head() #display first 5 rows
```

```python
data.shape # diamension of the dataframe
```

```python
data.info() #information about the dataframe
```

```python
data.columns  #display column names
```

```python
data.drop(data.columns[0],axis = 1, inplace = True)  #removing the unnamed column
data.head()
```

```python
data.isnull().sum()   #checking for null values
```

```python
data.duplicated().sum()  # checking for duplicates
```

```python
data.drop_duplicates()  #removing duplicates
```

```python
# converting continuous variables into high moderate and low
```

```python
data['DSepal.Length'] = pd.cut(data['Sepal.Length'],3,labels = ['low','moderate','high'])
data['DSepal.Width'] = pd.cut(data['Sepal.Width'],3, labels = ['low','moderate','high'])
data['DPetal.Length'] = pd.cut(data['Petal.Length'],3,labels = ['low','moderate','high'])
data['dPetal.Width'] = pd.cut(data['Petal.Width'],3,labels = ['low','moderate','high'])
data.head()
```

```python
#dropping the continuous data
```

```python
data.drop(['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width'], axis = 1, inplace = True)
```
